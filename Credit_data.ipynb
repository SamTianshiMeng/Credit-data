```python
# Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.tree import export_text

# Load CreditData.csv
credit_data = pd.read_csv('/content/CreditData.csv')


from sklearn import preprocessing


lbl_en = preprocessing.LabelEncoder()

# list the categorical features
categorical_features = ['Status', 'history', 'Purpose',  'Savings','employment', 'sex', 'guarantors',
       'Property', 'OtherInstallment', 'Housing','Job', 'Telephone', 'foreign', 'Approved']

credit_data[categorical_features] = credit_data[categorical_features].apply(lbl_en.fit_transform)


# Sort the dataset
credit_data.sort_values("amount", axis = 0,inplace = True, na_position ='last')

X_credit = credit_data.iloc[:,0:20].values
y_credit = credit_data.iloc[:,-1].values

from sklearn.model_selection import train_test_split

# Divide dataset into 80:20 ratio
X_train_credit, X_test_credit, y_train_credit, y_test_credit = train_test_split(X_credit, y_credit, test_size=0.20, random_state=42)


# import decision tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree

decisionTree_cred = tree.DecisionTreeClassifier(random_state=0)
dt_cred_model = decisionTree_cred.fit(X_train_credit,y_train_credit)
tree.plot_tree(decisionTree_cred)
y_pred = dt_cred_model.predict(X_test_credit)


# Import the scores
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
print("Model performance")
print("Accuracy Score: ",accuracy_score(y_test_credit, y_pred)*100 ,"%")
print("precision Score: ",precision_score(y_test_credit, y_pred)*100,"%")
print("recall Score: ",recall_score(y_test_credit, y_pred)*100,"%")


```
